```python
"""
V7 (SKU-bazlı seçim) — Tüm SKU'larda EXOG × Y varyantlarını dener, en iyisini seçer,
6 aylık tahmin + PI80/PI95 üretir, stok tükeniş olasılıklarını hesaplar ve sipariş önerir.
Ayrıca her SKU için aday tabloyu konsola yazar, 6m ve 3m grafiklerini çizer ve PNG olarak kaydeder.

Girdi:
  - panel_sales_orders_stock.csv (kolonlar: sku, ds, y, orders, stock)
  - sku_config.csv (kolonlar: sku, T_CHECK, H_COVER, q_target, lead_time_mo, MOQ, lot_size)

Çıktı:
  - ./v7_select/forecasts/<sku>.csv
  - ./v7_select/plots/<sku>_6m.png, <sku>_3m.png
  - Konsol: aday kombinasyonlar ve seçilen yöntem özeti
"""

import pandas as pd, numpy as np, json, math, warnings, sys
from pathlib import Path
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import ParameterGrid

try:
    from xgboost import XGBRegressor
    HAS_XGB = True
except Exception:
    HAS_XGB = False

warnings.filterwarnings("ignore")

PANEL_PATH = "mnt/data/panel_sales_orders_stock.csv"
CFG_PATH   = "mnt/data/sku_config.csv"
OUT_DIR    = Path("mnt/data/v7_select")
(OUT_DIR/"forecasts").mkdir(parents=True, exist_ok=True)
(OUT_DIR/"plots").mkdir(parents=True, exist_ok=True)

rng_global = np.random.default_rng(42)

# -------- Utils ----------
def ensure_ms_freq(df):
    d = df.copy().sort_values("ds")
    d["ds"] = pd.to_datetime(d["ds"]).dt.to_period("M").dt.to_timestamp(how="start")
    d = d.drop_duplicates(["ds"])
    rng = pd.date_range(d["ds"].min(), d["ds"].max(), freq="MS")
    dd = pd.DataFrame({"ds": rng}).merge(d, on="ds", how="left")
    for c in ["y","orders"]:
        if c in dd.columns:
            dd[c] = pd.to_numeric(dd[c], errors="coerce").fillna(0.0).clip(lower=0)
    if "stock" in dd.columns:
        dd["stock"] = pd.to_numeric(dd["stock"], errors="coerce").ffill().fillna(0.0).clip(lower=0)
    return dd

def add_cal_and_lags(df):
    d = df.copy().sort_values("ds")
    d["month"] = d["ds"].dt.month
    d["year"]  = d["ds"].dt.year
    d["orders_ratio"] = d["orders"] / d["stock"].replace(0,np.nan)
    d["orders_ratio"] = d["orders_ratio"].replace([np.inf,-np.inf], np.nan).fillna(0.0)
    for c in ["y","orders","stock"]:
        d[f"{c}_lag1"] = d[c].shift(1)
        d[f"{c}_lag3"] = d[c].shift(3)
    lag_cols = [c for c in d.columns if c.endswith("_lag1") or c.endswith("_lag3")]
    d[lag_cols] = d[lag_cols].fillna(0.0)
    return d

def seasonal_naive(series, steps):
    s = series.values
    if len(s) >= 12:
        patt = s[-12:]
        reps = int(np.ceil(steps/12))
        fc = np.tile(patt, reps)[:steps]
    else:
        fc = np.repeat(s[-1] if len(s)>0 else 0.0, steps)
    return fc

def moving_avg(series, steps, window=3):
    base = series.rolling(window=window, min_periods=1).mean()
    last = float(base.iloc[-1]) if len(base)>0 else 0.0
    return np.repeat(max(0.0,last), steps)

def last_value(series, steps):
    last = float(series.iloc[-1]) if len(series)>0 else 0.0
    return np.repeat(max(0.0,last), steps)

def inverse_mae_weights(y_true, preds):
    maes = {}
    for k, yhat in preds.items():
        if len(y_true) == len(yhat) and len(y_true)>0:
            err = np.abs(y_true - yhat).mean()
        else:
            err = np.inf
        maes[k] = err
    inv = {k: (0.0 if (not np.isfinite(v) or v<=0) else 1.0/v) for k,v in maes.items()}
    s = sum(inv.values())
    if s == 0:
        K = len(preds) if len(preds)>0 else 1
        return {k: 1.0/K for k in preds}
    return {k: v/s for k,v in inv.items()}

def mae_rmse_mape(y_true, y_pred):
    y_true = np.asarray(y_true, float); y_pred = np.asarray(y_pred, float)
    mae  = np.mean(np.abs(y_true - y_pred))
    rmse = np.sqrt(np.mean((y_true - y_pred)**2))
    denom = np.where(y_true==0, 1.0, y_true)
    mape = np.mean(np.abs((y_true - y_pred)/denom))*100.0
    return mae, rmse, mape

def exog_build(df_hist, col, steps, method, val_h=6):
    s = df_hist.set_index("ds")[col].astype(float)
    if method == "last":
        return last_value(s, steps), {"last":1.0}
    if method == "ma3":
        return moving_avg(s, steps, 3), {"ma3":1.0}
    if method == "snaive12":
        return seasonal_naive(s, steps), {"snaive":1.0}
    # inverse over three
    val_h = min(val_h, max(1, len(s)//4))
    train = s.iloc[:-val_h] if len(s)>val_h else s
    val   = s.iloc[-val_h:] if len(s)>val_h else s.iloc[-1:]
    preds_val = {
        "snaive": seasonal_naive(train, len(val)),
        "last":   last_value(train, len(val)),
        "ma3":    moving_avg(train, len(val), 3),
    }
    w = inverse_mae_weights(val.values, preds_val)
    preds_fut = {
        "snaive": seasonal_naive(s, steps),
        "last":   last_value(s, steps),
        "ma3":    moving_avg(s, steps, 3),
    }
    mix = np.zeros(steps, float)
    for k, arr in preds_fut.items():
        mix += w.get(k,0.0) * arr
    return mix, w

def time_series_splits(n, n_splits=3, min_train=24, test_h=3):
    if n < (min_train + n_splits*test_h):
        n_splits = max(1, (n - min_train)//max(1,test_h))
    if n_splits <= 0:
        n_splits = 1
    for i in range(n_splits):
        tr_end = min_train + i*test_h
        te_end = tr_end + test_h
        if te_end > n: break
        tr_idx = np.arange(0, tr_end)
        te_idx = np.arange(tr_end, te_end)
        yield tr_idx, te_idx

def fit_rf_with_rocv(X, y):
    grid = {
        "n_estimators": [300, 500],
        "max_depth": [None, 8, 12],
        "min_samples_split": [2, 5],
        "min_samples_leaf": [1, 2],
    }
    best_params, best_mae = None, np.inf
    splits = list(time_series_splits(len(y), n_splits=3, min_train=min(24, max(12, len(y)//3)), test_h=3))
    if len(splits) == 0:
        splits = [(np.arange(0, max(1,len(y)-3)), np.arange(max(1,len(y)-3), len(y)))]
    for params in ParameterGrid(grid):
        maes = []
        for tr_idx, te_idx in splits:
            rf = RandomForestRegressor(random_state=42, n_jobs=-1, **params)
            rf.fit(X[tr_idx], y[tr_idx])
            yhat = rf.predict(X[te_idx])
            maes.append(np.mean(np.abs(y[te_idx]-yhat)))
        m = float(np.mean(maes))
        if m < best_mae:
            best_mae = m; best_params = params
    rf_final = RandomForestRegressor(random_state=42, n_jobs=-1, **best_params).fit(X, y)
    return rf_final, best_params, best_mae

def fit_xgb_with_rocv(X, y):
    if not HAS_XGB:
        return None, None, np.inf
    grid = {
        "n_estimators": [400, 700],
        "learning_rate": [0.05, 0.1],
        "max_depth": [3, 4],
        "subsample": [0.8, 1.0],
        "colsample_bytree": [0.8, 1.0],
        "reg_lambda": [1.0, 2.0],
    }
    best_params, best_mae = None, np.inf
    splits = list(time_series_splits(len(y), n_splits=3, min_train=min(24, max(12, len(y)//3)), test_h=3))
    if len(splits) == 0:
        splits = [(np.arange(0, max(1,len(y)-3)), np.arange(max(1,len(y)-3), len(y)))]
    for params in ParameterGrid(grid):
        maes = []
        for tr_idx, te_idx in splits:
            xgb = XGBRegressor(random_state=42, **params)
            xgb.fit(X[tr_idx], y[tr_idx], verbose=False)
            yhat = xgb.predict(X[te_idx])
            maes.append(np.mean(np.abs(y[te_idx]-yhat)))
        m = float(np.mean(maes))
        if m < best_mae:
            best_mae = m; best_params = params
    xgb_final = XGBRegressor(random_state=42, **best_params).fit(X, y, verbose=False)
    return xgb_final, best_params, best_mae

def smoothed_bootstrap_intervals(yhat_series, residuals, B=800, rng=None):
    rng = rng or rng_global
    res = np.asarray(residuals, float); res = res[np.isfinite(res)]
    n = res.size
    mad = np.median(np.abs(res - np.median(res))) if n>0 else 0.0
    b_lap = (mad/np.sqrt(2)) if mad>0 else (np.std(res) if n>1 else 1.0)
    if n == 0:
        noise = rng.laplace(0.0, max(1e-3, b_lap), size=(len(yhat_series), B))
    else:
        idx = rng.integers(0, n, size=(len(yhat_series), B))
        base = res[idx]; jitter = rng.laplace(0.0, max(0.25*b_lap, 1e-6), size=(len(yhat_series), B))
        noise = base + jitter
    sims = yhat_series.reshape(-1,1) + noise
    lo80 = np.nanquantile(sims, 0.10, axis=1); hi80 = np.nanquantile(sims, 0.90, axis=1)
    lo95 = np.nanquantile(sims, 0.025, axis=1); hi95 = np.nanquantile(sims, 0.975, axis=1)
    return sims, lo80, hi80, lo95, hi95

def stockout_stats(start_stock, sims):
    H, B = sims.shape
    cums = np.cumsum(np.maximum(0.0, sims), axis=0)
    crossed = (cums >= start_stock)
    p3 = float(np.mean(crossed[min(3,H)-1,:])) if H>=1 else 0.0
    p6 = float(np.mean(crossed[min(6,H)-1,:])) if H>=1 else 0.0
    first_cross = np.argmax(crossed, axis=0) + 1
    never = ~crossed.any(axis=0)
    first_cross[never] = H+1
    e_t = float(np.mean(first_cross))
    return p3, p6, e_t

def round_moq_lot(q, moq=0, lot=1):
    q = max(0.0, q)
    if lot <= 0: lot = 1
    if q > 0 and q < moq: q = moq
    q = math.ceil(q / lot) * lot
    return float(q)

def run():
    panel = pd.read_csv(PANEL_PATH, parse_dates=["ds"]).sort_values(["sku","ds"]).reset_index(drop=True)
    cfg   = pd.read_csv(CFG_PATH)
    for col, default in {"T_CHECK":3,"H_COVER":6,"q_target":0.50,"lead_time_mo":0,"MOQ":0,"lot_size":1}.items():
        if col not in cfg.columns: cfg[col] = default

    FEATS = ["orders","stock","orders_lag1","orders_lag3","stock_lag1","stock_lag3","y_lag1","y_lag3","orders_ratio","month","year"]

    for sku, df in panel.groupby("sku", sort=False):
        row_cfg = cfg[cfg["sku"].astype(str)==str(sku)]
        if row_cfg.empty:
            T_CHECK=3; H_COVER=6; q_target=0.50; lead=0; moq=0; lot=1
        else:
            T_CHECK = int(row_cfg["T_CHECK"].iloc[0])
            H_COVER = int(row_cfg["H_COVER"].iloc[0])
            q_target= float(row_cfg["q_target"].iloc[0])
            lead    = int(row_cfg["lead_time_mo"].iloc[0])
            moq     = int(row_cfg["MOQ"].iloc[0])
            lot     = int(row_cfg["lot_size"].iloc[0])

        d = df[["ds","y","orders","stock"]].copy()
        d = ensure_ms_freq(d).fillna(0.0)
        feat = add_cal_and_lags(d)

        # split
        if len(feat) < 24:
            tr = feat.iloc[:-3]; va = feat.iloc[-3:]
        else:
            tr = feat.iloc[:-6]; va = feat.iloc[-6:]

        X_tr = tr[FEATS].to_numpy(); y_tr = tr["y"].to_numpy()

        # Fit Y-variants
        rf, rf_params, rf_mae = fit_rf_with_rocv(X_tr, y_tr)
        if HAS_XGB:
            xgb, xgb_params, xgb_mae = fit_xgb_with_rocv(X_tr, y_tr)
        else:
            xgb, xgb_params, xgb_mae = None, None, np.inf

        y_variants = {"RF": (rf, rf_mae, rf_params)}
        if HAS_XGB:
            y_variants["XGB"] = (xgb, xgb_mae, xgb_params)

        # Build EXOG candidates
        H_FUT = max(6, H_COVER+lead)
        fut_idx = pd.date_range(d["ds"].max()+pd.offsets.MonthBegin(1), periods=H_FUT, freq="MS")

        exog_methods = ["last","ma3","snaive12","inverse3"]
        exog_candidates = {}
        exog_weights_dbg = {}
        for m in exog_methods:
            fut = pd.DataFrame({"ds": fut_idx})
            for col in ["orders","stock"]:
                arr, w = exog_build(d[["ds",col]], col, H_FUT, m, val_h=6)
                fut[col] = np.maximum(0.0, arr)
                exog_weights_dbg[(m,col)] = w
            exog_candidates[m] = fut.copy()

        # Evaluate candidates on validation (1-step ahead)
        rows_cand = []
        for exname, _ in exog_candidates.items():
            for yname, (mdl, mae_cv, params) in y_variants.items():
                resids = []
                rolling = tr.copy()
                for _, row in va.iterrows():
                    mdl2 = type(mdl)(**mdl.get_params()) if hasattr(mdl, "get_params") else mdl
                    if yname == "XGB" and HAS_XGB:
                        mdl2 = XGBRegressor(**mdl.get_params())
                        mdl2.fit(rolling[FEATS].to_numpy(), rolling["y"].to_numpy(), verbose=False)
                    else:
                        mdl2.fit(rolling[FEATS].to_numpy(), rolling["y"].to_numpy())
                    ph = float(mdl2.predict(row[FEATS].to_numpy().reshape(1,-1))[0])
                    resids.append(row["y"] - ph)
                    rolling = pd.concat([rolling, row.to_frame().T], ignore_index=True)
                if len(resids)==0:
                    val_mae, val_rmse, val_mape = np.inf, np.inf, np.inf
                else:
                    y_val = va["y"].to_numpy()
                    yhat_val = y_val - np.array(resids)
                    val_mae, val_rmse, val_mape = mae_rmse_mape(y_val, yhat_val)
                rows_cand.append({
                    "SKU": sku, "Exog": exname, "Y": yname,
                    "VAL_MAE": val_mae, "VAL_RMSE": val_rmse, "VAL_MAPE": val_mape,
                    "CV_MAE": mae_cv
                })
        cand_df = pd.DataFrame(rows_cand).sort_values("VAL_MAE").reset_index(drop=True)

        # Print candidates to console
        print(f"\n==== {sku} — Aday Kombinasyonlar (VAL sıralı) ====")
        with pd.option_context("display.max_rows", None, "display.max_columns", None, "display.width", 120):
            print(cand_df)

        # Select best combo
        best_row = cand_df.iloc[0]
        best_ex = best_row["Exog"]; best_y = best_row["Y"]

        # Refit chosen model on full data
        model = rf if best_y=="RF" else (xgb if HAS_XGB else rf)
        if best_y=="XGB" and HAS_XGB:
            model = XGBRegressor(**model.get_params())
            model.fit(feat[FEATS].to_numpy(), feat["y"].to_numpy(), verbose=False)
        else:
            model = type(model)(**model.get_params())
            model.fit(feat[FEATS].to_numpy(), feat["y"].to_numpy())

        # Build chosen exog future
        ex_fut = exog_candidates[best_ex].copy()

        # Recursive 6m forecast with chosen combo
        H = 6
        preds = []; full = feat.copy()
        for ds in pd.date_range(d["ds"].max()+pd.offsets.MonthBegin(1), periods=H, freq="MS"):
            full = full[~(full["ds"]==ds)]
            new_row = {"ds": ds, "y": 0.0,
                       "orders": float(ex_fut.loc[ex_fut["ds"]==ds,"orders"].values[0]),
                       "stock":  float(ex_fut.loc[ex_fut["ds"]==ds,"stock"].values[0])}
            full = pd.concat([full, pd.DataFrame([new_row])], ignore_index=True).sort_values("ds")
            full = add_cal_and_lags(full)
            X = full.loc[full["ds"]==ds, FEATS].to_numpy()
            yhat = float(model.predict(X)[0])
            preds.append({"ds": ds, "yhat": max(0.0, yhat)})
            full.loc[full["ds"]==ds, "y"] = yhat
        pred_df = pd.DataFrame(preds)

        # Residual pool from validation (best_y)
        resids = []
        rolling = tr.copy()
        for _, row in va.iterrows():
            mdl2 = type(model)(**model.get_params()) if best_y!="XGB" else XGBRegressor(**model.get_params())
            if best_y=="XGB" and HAS_XGB:
                mdl2.fit(rolling[FEATS].to_numpy(), rolling["y"].to_numpy(), verbose=False)
            else:
                mdl2.fit(rolling[FEATS].to_numpy(), rolling["y"].to_numpy())
            ph = float(mdl2.predict(row[FEATS].to_numpy().reshape(1,-1))[0])
            resids.append(row["y"] - ph)
            rolling = pd.concat([rolling, row.to_frame().T], ignore_index=True)
        resids = np.array(resids, dtype=float)

        sims, lo80, hi80, lo95, hi95 = smoothed_bootstrap_intervals(pred_df["yhat"].to_numpy(), resids, B=800, rng=rng_global)
        fc = pred_df.copy()
        fc["pi80_lo"] = np.maximum(0.0, lo80); fc["pi80_hi"] = np.maximum(0.0, hi80)
        fc["pi95_lo"] = np.maximum(0.0, lo95); fc["pi95_hi"] = np.maximum(0.0, hi95)

        # Stockout + policy
        start_stock = float(d["stock"].iloc[-1]) if len(d)>0 else 0.0
        p3, p6, e_t = stockout_stats(start_stock, sims)

        total_cov = sims[:min(H, H_COVER), :].sum(axis=0)
        q_total = float(np.nanquantile(total_cov, q_target))
        raw_order = max(0.0, q_total - start_stock)
        order_qty = round_moq_lot(raw_order, moq=moq, lot=lot)

        # Save forecast CSV
        out_path = OUT_DIR/"forecasts"/f"{sku}.csv"
        fc.to_csv(out_path, index=False, date_format="%Y-%m-%d")

        # ---- Print selection summary ----
        print(f"\n=== {sku} — Seçilen En İyi Kombinasyon ===")
        print(f"Exog: {best_ex} | Y: {best_y}")
        print(f"VAL_MAE={best_row['VAL_MAE']:.2f} | VAL_RMSE={best_row['VAL_RMSE']:.2f} | VAL_MAPE={best_row['VAL_MAPE']:.2f}")
        print(f"Stockout P<=3m={p3:.3f} P<=6m={p6:.3f} | E[T]={e_t:.2f} ay")
        print(f"Policy: cover {H_COVER}m @ q={q_target} | MOQ={moq} lot={lot}")
        print(f"Önerilen Sipariş: {order_qty:.2f} (ham={raw_order:.2f}, stok={start_stock:.2f})")

        # ---- Plots ----
        hist = d.copy()
        hist12 = hist[hist["ds"] >= hist["ds"].max() - pd.DateOffset(months=11)]
        # 6m
        plt.figure(figsize=(10,5))
        plt.plot(hist12["ds"], hist12["y"], label="Gerçek (son 12 ay)")
        plt.plot(fc["ds"], fc["yhat"], label="Tahmin (6 ay)")
        plt.fill_between(fc["ds"], fc["pi80_lo"], fc["pi80_hi"], alpha=0.2, label="PI80")
        plt.fill_between(fc["ds"], fc["pi95_lo"], fc["pi95_hi"], alpha=0.1, label="PI95")
        plt.title(f"{sku} — 6 Aylık Tahmin (Seçim: {best_ex} + {best_y})")
        plt.xlabel("Tarih"); plt.ylabel("Satış"); plt.legend(); plt.tight_layout()
        plt.savefig(OUT_DIR/"plots"/f"{sku}_6m.png", dpi=120)
        plt.show()

        # 3m zoom
        plt.figure(figsize=(10,5))
        plt.plot(hist12["ds"], hist12["y"], label="Gerçek (son 12 ay)")
        fc3 = fc.iloc[:3]
        plt.plot(fc3["ds"], fc3["yhat"], label="Tahmin (3 ay)")
        plt.fill_between(fc3["ds"], fc3["pi80_lo"], fc3["pi80_hi"], alpha=0.2, label="PI80")
        plt.fill_between(fc3["ds"], fc3["pi95_lo"], fc3["pi95_hi"], alpha=0.1, label="PI95")
        plt.title(f"{sku} — 3 Aylık Tahmin (Seçim: {best_ex} + {best_y})")
        plt.xlabel("Tarih"); plt.ylabel("Satış"); plt.legend(); plt.tight_layout()
        plt.savefig(OUT_DIR/"plots"/f"{sku}_3m.png", dpi=120)
        plt.show()

    print("\nTamamlandı. Çıktılar:", str(OUT_DIR))

if __name__ == "__main__":
    run()
```
