```python
# -*- coding: utf-8 -*-
"""
V7_full — V6 ile aynı kapsam: tüm EXOG kombinasyonları + Y varyantları,
SKU-bazında 6m VAL + 6m TEST bölmesi, NNLS/inverse/adaptive ağırlıklar, PI80/95,
stockout olasılıkları ve sipariş önerisi.

Girdiler:
  - panel_sales_orders_stock.csv  (sku, ds, y, orders, stock)
  - sku_config.csv                (sku, T_CHECK, H_COVER, q_target, lead_time_mo, MOQ, lot_size)

Çıktılar:
  - ./v7_full/forecasts/<sku>__<exog>__<yvar>.csv
  - ./v7_full/plots/<sku>__<exog>__<yvar>_6m.png ve _3m.png
  - Konsol: VAL & TEST özet tabloları, seçilen EXOG, Y-ENS ağırlıkları, sipariş önerisi
"""

import warnings, math
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pathlib import Path

# Y-öğreniciler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import ParameterGrid

try:
    from xgboost import XGBRegressor
    HAS_XGB = True
except Exception:
    HAS_XGB = False

# Zaman serisi modelleri (exog klasik)
from prophet import Prophet
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# ------------ Dosya yolları ------------
PANEL_PATH = "panel_sales_orders_stock.csv"
CFG_PATH   = "sku_config.csv"
OUT_DIR    = Path("v7_full")
(OUT_DIR/"forecasts").mkdir(parents=True, exist_ok=True)
(OUT_DIR/"plots").mkdir(parents=True, exist_ok=True)

rng_global = np.random.default_rng(42)

# ------------ Yardımcılar ------------
def ensure_ms_freq(df):
    d = df.copy().sort_values("ds")
    d["ds"] = pd.to_datetime(d["ds"]).dt.to_period("M").dt.to_timestamp(how="start")
    d = d.drop_duplicates(["ds"])
    rng = pd.date_range(d["ds"].min(), d["ds"].max(), freq="MS")
    dd = pd.DataFrame({"ds": rng}).merge(d, on="ds", how="left")
    for c in ["y","orders"]:
        if c in dd.columns:
            dd[c] = pd.to_numeric(dd[c], errors="coerce").fillna(0.0).clip(lower=0)
    if "stock" in dd.columns:
        dd["stock"] = pd.to_numeric(dd["stock"], errors="coerce").ffill().fillna(0.0).clip(lower=0)
    return dd

def add_calendar_lags(df):
    d = df.copy().sort_values("ds")
    d["month"] = d["ds"].dt.month
    d["year"]  = d["ds"].dt.year
    d["orders_ratio"] = d["orders"] / d["stock"].replace(0,np.nan)
    d["orders_ratio"] = d["orders_ratio"].replace([np.inf,-np.inf], np.nan).fillna(0.0)
    for c in ["y","orders","stock"]:
        d[f"{c}_lag1"] = d[c].shift(1)
        d[f"{c}_lag3"] = d[c].shift(3)
    for c in [col for col in d.columns if c.endswith("_lag1") or col.endswith("_lag3")]:
        d[c] = d[c].fillna(0.0)
    return d

def mae_rmse_mape(y_true, y_pred):
    y_true = np.asarray(y_true, float); y_pred = np.asarray(y_pred, float)
    mae  = np.mean(np.abs(y_true - y_pred))
    rmse = np.sqrt(np.mean((y_true - y_pred)**2))
    denom = np.where(y_true==0, 1.0, y_true)
    mape = np.mean(np.abs((y_true - y_pred)/denom))*100.0
    return mae, rmse, mape

def split_train_val_test(d, val_h=6, test_h=6):
    d = d.sort_values("ds").reset_index(drop=True)
    test_end = d["ds"].max()
    test_start = test_end - pd.DateOffset(months=test_h-1)
    val_end = test_start - pd.DateOffset(months=1)
    val_start = val_end - pd.DateOffset(months=val_h-1)
    train = d[d["ds"] < val_start]
    val   = d[(d["ds"] >= val_start) & (d["ds"] <= val_end)]
    test  = d[(d["ds"] >= test_start) & (d["ds"] <= test_end)]
    return train.reset_index(drop=True), val.reset_index(drop=True), test.reset_index(drop=True), (val_start, val_end, test_start, test_end)

# ------------ EXOG klasik (Prophet/SARIMA/ETS) ------------
def fit_prophet(train_df, value_col):
    m = Prophet(yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False)
    tmp = train_df.rename(columns={value_col:"y"})[["ds","y"]]
    m.fit(tmp)
    return m

def fc_prophet(model, steps):
    fut = model.make_future_dataframe(periods=steps, freq="MS")
    fc  = model.predict(fut)[["ds","yhat"]].tail(steps).rename(columns={"yhat":"yhat"})
    return fc

def sarima_fit_best(y, p_range=(0,3), q_range=(0,3), P_range=(0,1), Q_range=(0,1)):
    best, best_aic = None, np.inf
    for p in range(p_range[0], p_range[1]+1):
        for q in range(q_range[0], q_range[1]+1):
            for P in range(P_range[0], P_range[1]+1):
                for Q in range(Q_range[0], Q_range[1]+1):
                    try:
                        r = SARIMAX(y, order=(p,1,q), seasonal_order=(P,1,Q,12),
                                    enforce_stationarity=False, enforce_invertibility=False).fit(disp=False)
                        if r.aic < best_aic:
                            best_aic = r.aic; best = ((p,1,q),(P,1,Q,12))
                    except Exception:
                        pass
    return best

def fit_sarima(train_df, value_col):
    y = train_df.set_index("ds")[value_col].astype(float)
    y.index = pd.DatetimeIndex(y.index, freq="MS")
    best = sarima_fit_best(y)
    if best is None: return None
    return SARIMAX(y, order=best[0], seasonal_order=best[1],
                   enforce_stationarity=False, enforce_invertibility=False).fit(disp=False)

def fc_sarima(model, steps, future_idx):
    pred = model.get_forecast(steps=steps).predicted_mean
    return pd.DataFrame({"ds": pd.DatetimeIndex(future_idx), "yhat": pred.values})

def fit_ets(train_df, value_col):
    y = train_df.set_index("ds")[value_col].astype(float)
    y.index = pd.DatetimeIndex(y.index, freq="MS")
    best_aic, best_model = np.inf, None
    for trend in ["add","mul",None]:
        for seasonal in ["add","mul",None]:
            for damp in [True, False]:
                try:
                    if seasonal is None:
                        mdl = ExponentialSmoothing(y, trend=trend, seasonal=None, damped_trend=damp).fit(optimized=True)
                    else:
                        mdl = ExponentialSmoothing(y, trend=trend, seasonal=seasonal, seasonal_periods=12, damped_trend=damp).fit(optimized=True)
                    aic = getattr(mdl, "aic", np.inf)
                    if aic < best_aic:
                        best_aic, best_model = aic, mdl
                except Exception:
                    continue
    if best_model is None:
        best_model = ExponentialSmoothing(y, trend="add", seasonal="add", seasonal_periods=12).fit(optimized=True)
    return best_model

def fc_ets(model, steps, future_idx):
    pred = model.forecast(steps)
    return pd.DataFrame({"ds": pd.DatetimeIndex(future_idx), "yhat": pred.values})

# ------------ Basit postprocess ------------
def smooth_causal_ma(s, window=3):
    x = pd.to_numeric(s, errors="coerce").ffill()
    return x.rolling(window=window, min_periods=1).mean().bfill()

def winsorize_series(s, lq=0.05, uq=0.95):
    x = pd.to_numeric(s, errors="coerce")
    lo = np.nanpercentile(x, lq*100)
    hi = np.nanpercentile(x, uq*100)
    return x.clip(lo, hi)

def nonneg(s): return pd.to_numeric(s, errors="coerce").clip(lower=0.0)

def _postprocess_exog(df):
    d = df.copy()
    for c in ["orders","stock"]:
        if c in d.columns:
            d[c] = smooth_causal_ma(d[c], window=3)
            d[c] = winsorize_series(d[c], 0.05, 0.95)
            d[c] = nonneg(d[c])
    return d

# ------------ EXOG üreticileri ------------
def exog_univar_inverse(df_all, future_idx, val_h=6, eps_prophet=0.05, test_start=None):
    out = pd.DataFrame({"ds": future_idx})
    for col in ["orders","stock"]:
        s = df_all[["ds",col]].dropna().sort_values("ds")
        if test_start is not None:
            s = s[s["ds"] < test_start]
        steps = len(future_idx)
        tail = s.tail(val_h+6) if len(s) > val_h+6 else s.copy()
        cut  = tail["ds"].max() - pd.DateOffset(months=val_h-1)
        train = s[s["ds"] < cut]
        val   = s[s["ds"] >= cut]
        maes = []
        parts = {}
        try:
            mp  = fit_prophet(train, col)
            fcp = fc_prophet(mp, len(val)).rename(columns={"yhat":"p_val"})
            fcp_f = fc_prophet(mp, steps).rename(columns={"yhat":"p"})
            maes.append(("p", mae_rmse_mape(val[col].values, fcp["p_val"].values)[0] if len(val)>0 else np.inf))
            parts["p"] = fcp_f[["ds","p"]]
        except Exception:
            parts["p"] = pd.DataFrame({"ds": future_idx, "p": np.nan})
            maes.append(("p", np.inf))
        try:
            ms  = fit_sarima(train, col)
            fcs = fc_sarima(ms, len(val), val["ds"].values).rename(columns={"yhat":"s_val"})
            fcs_f = fc_sarima(ms, steps, future_idx).rename(columns={"yhat":"s"})
            maes.append(("s", mae_rmse_mape(val[col].values, fcs["s_val"].values)[0] if len(val)>0 else np.inf))
            parts["s"] = fcs_f[["ds","s"]]
        except Exception:
            parts["s"] = pd.DataFrame({"ds": future_idx, "s": np.nan})
            maes.append(("s", np.inf))
        try:
            me  = fit_ets(train, col)
            fce = fc_ets(me, len(val), val["ds"].values).rename(columns={"yhat":"e_val"})
            fce_f = fc_ets(me, steps, future_idx).rename(columns={"yhat":"e"})
            maes.append(("e", mae_rmse_mape(val[col].values, fce["e_val"].values)[0] if len(val)>0 else np.inf))
            parts["e"] = fce_f[["ds","e"]]
        except Exception:
            parts["e"] = pd.DataFrame({"ds": future_idx, "e": np.nan})
            maes.append(("e", np.inf))

        arr = np.array([m[1] for m in maes], float)
        inv = 1.0 / np.where((~np.isfinite(arr)) | (arr<=0), np.inf, arr)
        if not np.isfinite(inv).any():
            w = np.array([0.60, 0.25, 0.15], float)
        else:
            w = inv / inv.sum()
            w[0] += eps_prophet
            w = w / w.sum()
        wp, ws, we = w
        tmp = pd.DataFrame({"ds": future_idx})
        for k in ["p","s","e"]:
            tmp = tmp.merge(parts[k], on="ds", how="left")
        tmp[col] = wp*tmp["p"] + ws*tmp["s"] + we*tmp["e"]
        out = out.merge(tmp[["ds",col]], on="ds", how="left")
    return _postprocess_exog(out)

EXOG_FEATS = ["lag1","lag3","month","year"]

def make_exog_feat(df_var, col):
    d = df_var[["ds", col]].sort_values("ds").copy()
    d["lag1"] = d[col].shift(1); d["lag3"] = d[col].shift(3)
    d["month"] = d["ds"].dt.month; d["year"] = d["ds"].dt.year
    return d

def train_exog_rf(df_var, col, cutoff):
    d = make_exog_feat(df_var[df_var["ds"] < cutoff], col).dropna()
    X = d[EXOG_FEATS]; y = d[col]
    rf = RandomForestRegressor(n_estimators=400, max_depth=8, min_samples_split=2, min_samples_leaf=1,
                               random_state=42, n_jobs=-1)
    rf.fit(X, y); return rf

def train_exog_xgb(df_var, col, cutoff):
    if not HAS_XGB:
        return None
    d = make_exog_feat(df_var[df_var["ds"] < cutoff], col).dropna()
    X = d[EXOG_FEATS].to_numpy(); y = d[col].to_numpy()
    xgb = XGBRegressor(n_estimators=500, learning_rate=0.08, max_depth=3,
                       subsample=0.9, colsample_bytree=0.9, reg_lambda=1.2, random_state=42)
    xgb.fit(X, y, verbose=False); return xgb

def recursive_forecast_exog(model, hist_df, col, start_ds, end_ds):
    future_idx = pd.date_range(start_ds, end_ds, freq="MS")
    full = hist_df[["ds", col]].sort_values("ds").copy()
    out = []
    for ds in future_idx:
        tmp = make_exog_feat(full, col)
        row = tmp[tmp["ds"] == ds][EXOG_FEATS]
        if row.empty:
            row = pd.DataFrame({
                "lag1":[full[col].iloc[-1] if len(full) else 0.0],
                "lag3":[full[col].iloc[-3] if len(full)>=3 else (full[col].iloc[-1] if len(full) else 0.0)],
                "month":[ds.month], "year":[ds.year]
            })
        yhat = float(model.predict(row.to_numpy())[0])
        out.append({"ds": ds, col: max(0.0, yhat)})
        full = pd.concat([full, pd.DataFrame({"ds":[ds], col:[yhat]})], ignore_index=True)
    return pd.DataFrame(out)

def build_exog_ml(df_all, future_idx, learner="rf", test_start=None):
    out = pd.DataFrame({"ds": future_idx})
    for col in ["orders","stock"]:
        s = df_all[["ds", col]].dropna().sort_values("ds")
        cutoff = test_start if test_start is not None else future_idx.min()
        if learner == "rf":
            mdl = train_exog_rf(s, col, cutoff=cutoff)
        else:
            mdl = train_exog_xgb(s, col, cutoff=cutoff)
        fc = recursive_forecast_exog(mdl, df_all[df_all["ds"] < cutoff][["ds", col]].copy(), col,
                                     start_ds=future_idx.min(), end_ds=future_idx.max())
        out = out.merge(fc, on="ds", how="left")
    return _postprocess_exog(out)

# ------------ EXOG stacking katmanları ------------
def nnls_weights(A, y, alpha=0.0):
    A = np.asarray(A, float); y = np.asarray(y, float)
    K = A.shape[1]
    H = A.T @ A + alpha*np.eye(K)
    b = A.T @ y
    try:
        w = np.linalg.solve(H, b)
    except np.linalg.LinAlgError:
        w = np.linalg.lstsq(H, b, rcond=None)[0]
    w = np.clip(w, 0.0, None)
    s = w.sum()
    if s > 0: w = w / s
    return w

def stack_exogs(exog_dict, future_idx, val_truth=None, mode="INV", alpha=0.0):
    names = list(exog_dict.keys())
    if mode in ("NNLS","ADAPTIVE_NNLS") and (val_truth is None or val_truth.empty):
        raise ValueError("NNLS için val_true gerekir.")
    fut = pd.DataFrame({"ds": future_idx})
    for col in ["orders","stock"]:
        mats = []
        for nm in names:
            mats.append(exog_dict[nm][col].to_numpy())
        A_fut = np.vstack(mats).T  # (H, K)

        if mode == "INV":
            maes = []
            for nm in names:
                if f"{col}_VAL" in exog_dict[nm].columns:
                    yhat_v = exog_dict[nm][f"{col}_VAL"].to_numpy()
                    ytrue  = val_truth[col].to_numpy()
                    maes.append(np.mean(np.abs(ytrue - yhat_v)))
                else:
                    maes.append(np.nan)
            arr = np.array(maes, float)
            inv = 1.0/np.where((~np.isfinite(arr))|(arr<=0), np.inf, arr)
            if not np.isfinite(inv).any():
                w = np.ones(len(names))/max(1,len(names))
            else:
                w = inv / inv.sum()
        elif mode in ("NNLS","ADAPTIVE_NNLS"):
            A_val = np.vstack([exog_dict[nm][f"{col}_VAL"].to_numpy() for nm in names]).T
            y_val = val_truth[col].to_numpy()
            w = nnls_weights(A_val, y_val, alpha=alpha)
        else:
            w = np.ones(len(names))/len(names)

        fut[col] = np.maximum(0.0, A_fut @ w)
    return _postprocess_exog(fut)

# ------------ Y-öğrenici HP arama (ROCV) ------------
FEATS_Y = ["orders","stock","orders_lag1","orders_lag3","stock_lag1","stock_lag3","y_lag1","y_lag3","orders_ratio","month","year"]

def time_series_splits(n, n_splits=3, min_train=24, test_h=3):
    if n < (min_train + n_splits*test_h):
        n_splits = max(1, (n - min_train)//max(1,test_h))
    if n_splits <= 0: n_splits = 1
    for i in range(n_splits):
        tr_end = min_train + i*test_h
        te_end = tr_end + test_h
        if te_end > n: break
        yield np.arange(0, tr_end), np.arange(tr_end, te_end)

def optimize_rf(X, y):
    grid = {
        "n_estimators":[300,500,800],
        "max_depth":[None,8,12],
        "min_samples_split":[2,5],
        "min_samples_leaf":[1,2],
    }
    best, best_mae = None, np.inf
    splits = list(time_series_splits(len(y), n_splits=3, min_train=min(24, max(12, len(y)//3)), test_h=3))
    if not splits: splits=[(np.arange(0,max(1,len(y)-3)), np.arange(max(1,len(y)-3), len(y)))]
    for p in ParameterGrid(grid):
        maes=[]
        for tr_idx, te_idx in splits:
            mdl=RandomForestRegressor(random_state=42, n_jobs=-1, **p)
            mdl.fit(X[tr_idx], y[tr_idx])
            ph = mdl.predict(X[te_idx])
            maes.append(np.mean(np.abs(y[te_idx]-ph)))
        m=float(np.mean(maes))
        if m < best_mae: best_mae, best = m, p
    mdl=RandomForestRegressor(random_state=42, n_jobs=-1, **best).fit(X, y)
    return mdl, best, best_mae

def optimize_xgb(X, y):
    if not HAS_XGB:
        return None, None, np.inf
    grid = {
        "n_estimators":[400,700],
        "learning_rate":[0.05,0.1],
        "max_depth":[3,4],
        "subsample":[0.8,1.0],
        "colsample_bytree":[0.8,1.0],
        "reg_lambda":[1.0,2.0],
    }
    best, best_mae = None, np.inf
    splits = list(time_series_splits(len(y), n_splits=3, min_train=min(24, max(12, len(y)//3)), test_h=3))
    if not splits: splits=[(np.arange(0,max(1,len(y)-3)), np.arange(max(1,len(y)-3), len(y)))]
    for p in ParameterGrid(grid):
        maes=[]
        for tr_idx, te_idx in splits:
            mdl=XGBRegressor(random_state=42, **p)
            mdl.fit(X[tr_idx], y[tr_idx], verbose=False)
            ph = mdl.predict(X[te_idx])
            maes.append(np.mean(np.abs(y[te_idx]-ph)))
        m=float(np.mean(maes))
        if m < best_mae: best_mae, best = m, p
    mdl=XGBRegressor(random_state=42, **best).fit(X, y, verbose=False)
    return mdl, best, best_mae

# ------------ Y recursive tahmin ------------
def recursive_forward_predict_y(model, x_cols, hist_df, future_exog, start_ds, end_ds):
    future_idx = pd.date_range(start_ds, end_ds, freq="MS")
    future_part = pd.DataFrame({"ds": future_idx}).merge(future_exog, on="ds", how="left")
    full = pd.concat([hist_df, future_part], ignore_index=True).sort_values("ds")

    preds=[]
    for ds in future_idx:
        tmp = add_calendar_lags(full.copy())
        row = tmp.loc[tmp["ds"]==ds]
        X = row[x_cols].replace([np.inf,-np.inf], np.nan).fillna(0.0).to_numpy()
        yhat = float(model.predict(X)[0])
        preds.append({"ds": ds, "yhat": max(0.0, yhat)})
        full.loc[full["ds"]==ds, "y"] = yhat
        for c in ["orders","stock"]:
            if c in full.columns:
                full[c] = smooth_causal_ma(full[c], window=3)
    return pd.DataFrame(preds), full.loc[full["ds"].isin(future_idx)].copy()

# ------------ Bootstrap PI + stockout ------------
def smoothed_bootstrap_intervals(yhat_series, residuals, B=800, rng=None):
    rng = rng or rng_global
    res = np.asarray(residuals, float); res = res[np.isfinite(res)]
    n = res.size
    mad = np.median(np.abs(res - np.median(res))) if n>0 else 0.0
    b_lap = (mad/np.sqrt(2)) if mad>0 else (np.std(res) if n>1 else 1.0)
    if n == 0:
        noise = rng.laplace(0.0, max(1e-3, b_lap), size=(len(yhat_series), B))
    else:
        idx = rng.integers(0, n, size=(len(yhat_series), B))
        base = res[idx]; jitter = rng.laplace(0.0, max(0.25*b_lap, 1e-6), size=(len(yhat_series), B))
        noise = base + jitter
    sims = yhat_series.reshape(-1,1) + noise
    lo80 = np.nanquantile(sims, 0.10, axis=1); hi80 = np.nanquantile(sims, 0.90, axis=1)
    lo95 = np.nanquantile(sims, 0.025, axis=1); hi95 = np.nanquantile(sims, 0.975, axis=1)
    return sims, lo80, hi80, lo95, hi95

def stockout_stats(start_stock, sims):
    H, B = sims.shape
    cums = np.cumsum(np.maximum(0.0, sims), axis=0)
    crossed = (cums >= start_stock)
    p3 = float(np.mean(crossed[min(3,H)-1,:])) if H>=1 else 0.0
    p6 = float(np.mean(crossed[min(6,H)-1,:])) if H>=1 else 0.0
    first_cross = np.argmax(crossed, axis=0) + 1
    never = ~crossed.any(axis=0)
    first_cross[never] = H+1
    e_t = float(np.mean(first_cross))
    return p3, p6, e_t

def round_moq_lot(q, moq=0, lot=1):
    q = max(0.0, q)
    if lot <= 0: lot = 1
    if q > 0 and q < moq: q = moq
    q = math.ceil(q / lot) * lot
    return float(q)

# ------------ Ana akış (SKU-bazında) ------------
def main():
    panel = pd.read_csv(PANEL_PATH, parse_dates=["ds"]).sort_values(["sku","ds"]).reset_index(drop=True)
    cfg   = pd.read_csv(CFG_PATH)
    for col, default in {"T_CHECK":3,"H_COVER":6,"q_target":0.50,"lead_time_mo":0,"MOQ":0,"lot_size":1}.items():
        if col not in cfg.columns: cfg[col] = default

    for sku, raw in panel.groupby("sku", sort=False):
        print("\n" + "="*90)
        print(f"SKU: {sku}")
        d = raw[["ds","y","orders","stock"]].copy()
        d = ensure_ms_freq(d)
        feat = add_calendar_lags(d)

        train, val, test, (VAL_START, VAL_END, TEST_START, TEST_END) = split_train_val_test(feat[["ds","y","orders","stock"]], 6, 6)
        print(f"VAL: {VAL_START.date()}..{VAL_END.date()} | TEST: {TEST_START.date()}..{TEST_END.date()}")

        # Y HP arama (train)
        train_feat = add_calendar_lags(train)
        X_tr = train_feat[FEATS_Y].to_numpy(); y_tr = train_feat["y"].to_numpy()
        rf_model, rf_params, rf_cv = optimize_rf(X_tr, y_tr)
        if HAS_XGB:
            xgb_model, xgb_params, xgb_cv = optimize_xgb(X_tr, y_tr)
        else:
            xgb_model, xgb_params, xgb_cv = None, None, np.inf
        print(f"ROCV Y — RF {rf_params} | {rf_cv:.2f}  ;  XGB {xgb_params} | {xgb_cv:.2f}")

        future_idx = pd.date_range(TEST_START, TEST_END, freq="MS")

        # ---- EXOG adayları ----
        exog_ens = exog_univar_inverse(d[["ds","orders","stock"]], future_idx, val_h=6, eps_prophet=0.05, test_start=TEST_START)
        exog_rf  = build_exog_ml(d[["ds","orders","stock"]], future_idx, learner="rf",  test_start=TEST_START)
        exog_xgb = build_exog_ml(d[["ds","orders","stock"]], future_idx, learner="xgb", test_start=TEST_START) if HAS_XGB else None

        def exog_single(method):
            out = pd.DataFrame({"ds": future_idx})
            for col in ["orders","stock"]:
                s = d[["ds",col]]
                full = s[s["ds"] < TEST_START]
                tail = full.tail(6+6) if len(full) > 12 else full
                cut  = tail["ds"].max() - pd.DateOffset(months=6-1)
                trn  = full[full["ds"] < cut]
                steps = len(future_idx)
                if method=="prophet":
                    m = fit_prophet(trn, col); fc=fc_prophet(m, steps).rename(columns={"yhat":col})
                elif method=="sarima":
                    m = fit_sarima(trn, col);   fc=fc_sarima(m, steps, future_idx).rename(columns={"yhat":col})
                else:
                    m = fit_ets(trn, col);      fc=fc_ets(m, steps, future_idx).rename(columns={"yhat":col})
                out = out.merge(fc[["ds",col]], on="ds", how="left")
            return _postprocess_exog(out)

        exog_prophet = exog_single("prophet")
        exog_sarima  = exog_single("sarima")
        exog_ets     = exog_single("ets")

        if exog_xgb is not None:
            exog_rf_xgb = _postprocess_exog(pd.DataFrame({"ds": future_idx}).merge(exog_rf, on="ds").merge(exog_xgb, on="ds", suffixes=("_rf","_xgb")))
            exog_rf_xgb = pd.DataFrame({
                "ds": future_idx,
                "orders": 0.5*exog_rf_xgb["orders_rf"] + 0.5*exog_rf_xgb["orders_xgb"],
                "stock":  0.5*exog_rf_xgb["stock_rf"]  + 0.5*exog_rf_xgb["stock_xgb"],
            })
            exog_rf_xgb = _postprocess_exog(exog_rf_xgb)
        else:
            exog_rf_xgb = None

        EXOGS_BASIC = {
            "Prophet": exog_prophet,
            "SARIMA":  exog_sarima,
            "ETS":     exog_ets,
            "ML-Exog RF": exog_rf,
            "ML-Exog XGB": exog_xgb,
            "Ensemble": exog_ens,
        }
        if exog_rf_xgb is not None:
            EXOGS_BASIC["ML-Exog RF+XGB"] = exog_rf_xgb

        val_idx = pd.date_range(VAL_START, VAL_END, freq="MS")
        val_truth = d[(d["ds"]>=VAL_START)&(d["ds"]<=VAL_END)][["ds","orders","stock"]].reset_index(drop=True)

        def exog_val_mae(exog_df):
            if exog_df is None or len(exog_df)==0 or val_truth.empty: return np.inf
            tmp = val_truth.merge(exog_df.rename(columns={"orders":"orders_hat","stock":"stock_hat"}), on="ds", how="left")
            mae_o = np.mean(np.abs(tmp["orders"]-tmp["orders_hat"]))
            mae_s = np.mean(np.abs(tmp["stock"] -tmp["stock_hat"]))
            return float((mae_o+mae_s)/2.0)

        exog_rank = []
        for name, tab in EXOGS_BASIC.items():
            if tab is None: continue
            exog_rank.append((name, exog_val_mae(tab)))
        exog_rank = sorted(exog_rank, key=lambda x: x[1])
        names_sorted = [n for n,_ in exog_rank]

        ALL_LIST = [n for n in ["Prophet","SARIMA","ETS","ML-Exog RF","ML-Exog XGB","ML-Exog RF+XGB","Ensemble"] if (n in EXOGS_BASIC and EXOGS_BASIC[n] is not None)]
        Top2 = names_sorted[:2] if len(names_sorted)>=2 else names_sorted
        Top3 = names_sorted[:3] if len(names_sorted)>=3 else names_sorted

        # ========= VAL merge + fallback PATCH =========
        def stack_from_names(name_list, mode="INV", alpha=0.0):
            pool = {nm: EXOGS_BASIC[nm].copy() for nm in name_list}
            for nm in name_list:
                df_ex = pool[nm]
                align_val = val_truth.merge(
                    df_ex.rename(columns={"orders":"orders_pred", "stock":"stock_pred"}),
                    on="ds", how="left"
                )
                orders_val = align_val["orders_pred"]
                stock_val  = align_val["stock_pred"]
                if orders_val.isna().all():
                    orders_val = align_val["orders"]
                if stock_val.isna().all():
                    stock_val  = align_val["stock"]
                pool[nm][ "orders_VAL"] = orders_val.to_numpy()
                pool[nm][ "stock_VAL"]  = stock_val.to_numpy()
                pool[nm]["VAL"] = (
                    np.mean(np.abs(align_val["orders"].to_numpy() - pool[nm]["orders_VAL"])) +
                    np.mean(np.abs(align_val["stock"].to_numpy()  - pool[nm]["stock_VAL"]))
                ) / 2.0
            return stack_exogs(pool, future_idx, val_truth=val_truth, mode=mode, alpha=alpha)
        # ===========================================================

        exog_all5_inv   = stack_from_names([nm for nm in ALL_LIST if nm!="Ensemble"], mode="INV")
        exog_top2_inv   = stack_from_names(Top2, mode="INV")
        exog_top3_inv   = stack_from_names(Top3, mode="INV")
        exog_all5_nnls  = stack_from_names([nm for nm in ALL_LIST if nm!="Ensemble"], mode="NNLS", alpha=0.0)
        exog_top2_nnls  = stack_from_names(Top2, mode="NNLS", alpha=0.0)
        exog_top3_nnls  = stack_from_names(Top3, mode="NNLS", alpha=0.0)
        exog_adapt5_nnls= stack_from_names([nm for nm in ALL_LIST if nm!="Ensemble"], mode="ADAPTIVE_NNLS", alpha=0.0)

        EXOGS = {
            "Prophet": exog_prophet,
            "SARIMA":  exog_sarima,
            "ETS":     exog_ets,
            "Ensemble": exog_ens,
            "ML-Exog RF": exog_rf,
            "ML-Exog XGB": exog_xgb,
            "ML-Exog RF+XGB": exog_rf_xgb,
            "All-5-INV": exog_all5_inv,
            "Top2-INV":  exog_top2_inv,
            "Top3-INV":  exog_top3_inv,
            "All-5-NNLS": exog_all5_nnls,
            "Top2-NNLS":  exog_top2_nnls,
            "Top3-NNLS":  exog_top3_nnls,
            "Adaptive5-NNLS": exog_adapt5_nnls,
        }

        # >>> VAL features: val'i tekrar merge ETME!
        val_feat = add_calendar_lags(val.copy())
        hist_min = add_calendar_lags(pd.concat([train, val], ignore_index=True))

        # ====================== YENİ — exog beslemeli VAL değerlendirme ======================
        def eval_combo_on_VAL(exog_tab, y_model):
            """
            VAL sırasında özellik üretirken orders/stock'u SEÇİLEN exog_tab'tan alır.
            - Train döneminde gerçek orders/stock kullanılır (rolling başlangıcı).
            - VAL aylarında orders/stock = exog tahminleri; y tahmin edilir ve lag'lara girer (recursive).
            """
            # 1) Rolling başlangıç: TRAIN gerçekleri
            rolling = train[["ds", "y", "orders", "stock"]].copy().sort_values("ds").reset_index(drop=True)

            # 2) VAL indeksini exog ile hizala
            val_idx  = pd.date_range(val["ds"].min(), val["ds"].max(), freq="MS")
            exog_val = exog_tab.merge(pd.DataFrame({"ds": val_idx}), on="ds", how="right")

            rows = []
            for _, r in val.iterrows():
                ds = r["ds"]

                # 3) Bu ayın exog orders/stock değerleri
                row_ex = exog_val[exog_val["ds"] == ds]
                o = float(row_ex["orders"].iloc[0]) if (not row_ex.empty and "orders" in row_ex) else np.nan
                s = float(row_ex["stock"].iloc[0])  if (not row_ex.empty and "stock"  in row_ex) else np.nan

                # Fallback: exog NaN ise son gözlenen değeri kullan
                if not np.isfinite(o):
                    o = float(rolling["orders"].iloc[-1]) if len(rolling) else 0.0
                if not np.isfinite(s):
                    s = float(rolling["stock"].iloc[-1]) if len(rolling) else 0.0

                # 4) Mevcut rolling (train + önceki VAL tahminleri) ile modeli yeniden fit et
                roll_feat = add_calendar_lags(rolling)
                X_fit = roll_feat[FEATS_Y].to_numpy()
                y_fit = roll_feat["y"].to_numpy()

                mdl = type(y_model)(**y_model.get_params()) if hasattr(y_model, "get_params") else y_model
                if HAS_XGB and isinstance(mdl, XGBRegressor):
                    mdl.fit(X_fit, y_fit, verbose=False)
                else:
                    mdl.fit(X_fit, y_fit)

                # 5) Bu ay için feature satırı: orders/stock exog'dan, y bilinmiyor (tahmin edilecek)
                row_now = pd.DataFrame([{"ds": ds, "y": np.nan, "orders": o, "stock": s}])
                tmp = pd.concat([rolling, row_now], ignore_index=True)

                feat_now = add_calendar_lags(tmp).iloc[[-1]]
                X_now = (
                    feat_now[FEATS_Y]
                    .replace([np.inf, -np.inf], np.nan)
                    .fillna(0.0)
                    .to_numpy()
                )

                # 6) Tahmini yap → rolling'e yaz (recursive)
                yhat = float(max(0.0, mdl.predict(X_now)[0]))
                rows.append({"ds": ds, "yhat": yhat, "y": r["y"]})

                row_now.loc[:, "y"] = yhat
                rolling = pd.concat([rolling, row_now], ignore_index=True)

            sim = pd.DataFrame(rows).sort_values("ds")
            return mae_rmse_mape(sim["y"].to_numpy(), sim["yhat"].to_numpy())
        # =====================================================================================

        rows_val = []
        Y_models = {"RF": rf_model}
        if HAS_XGB: Y_models["XGB"] = xgb_model

        def yens_weights_on_val():
            maes = {}
            for yn, mdl in Y_models.items():
                m,_,_ = eval_combo_on_VAL(exog_ens, mdl)  # EXOG sabit: ens
                maes[yn]=m
            inv = {k:(0 if (v<=0 or not np.isfinite(v)) else 1.0/v) for k,v in maes.items()}
            s = sum(inv.values()) if inv else 0.0
            if s<=0: return {"RF":1.0} if not HAS_XGB else {"RF":0.5,"XGB":0.5}
            return {k:v/s for k,v in inv.items()}

        yens_w = yens_weights_on_val()

        for ex_name, ex_tab in EXOGS.items():
            if ex_tab is None: continue
            for yn, mdl in Y_models.items():
                m, r, p = eval_combo_on_VAL(ex_tab, mdl)
                rows_val.append({"Exog": ex_name, "Y": yn, "VAL_MAE": m, "VAL_RMSE": r, "VAL_MAPE": p})
            if HAS_XGB:
                m_rf, _, _  = eval_combo_on_VAL(ex_tab, rf_model)
                m_xgb, _, _ = eval_combo_on_VAL(ex_tab, xgb_model)
                m_yens = yens_w["RF"]*m_rf + yens_w["XGB"]*m_xgb
                rows_val.append({"Exog": ex_name, "Y": "Y-ENS", "VAL_MAE": m_yens, "VAL_RMSE": np.nan, "VAL_MAPE": np.nan})
        val_table = pd.DataFrame(rows_val).sort_values(["VAL_MAE","Exog","Y"]).reset_index(drop=True)

        print("\n=== VAL Exog Selection (all methods) ===")
        print(val_table.to_string(index=False))

        pick = val_table[val_table["Y"]=="Y-ENS"].sort_values("VAL_MAE").iloc[0] if HAS_XGB else val_table.sort_values("VAL_MAE").iloc[0]
        best_exog_name = pick["Exog"]
        print(f"\nSelected Exog by Y-ENS VAL MAE: {best_exog_name}")

        exog_selected = EXOGS[best_exog_name]
        hist_min2 = add_calendar_lags(pd.concat([train, val], ignore_index=True))
        full_feat = add_calendar_lags(pd.concat([train, val], ignore_index=True))
        X_full = full_feat[FEATS_Y].to_numpy(); y_full = full_feat["y"].to_numpy()

        rf_fit = RandomForestRegressor(**rf_model.get_params()).fit(X_full, y_full)
        if HAS_XGB:
            xgb_fit = XGBRegressor(**xgb_model.get_params()).fit(X_full, y_full, verbose=False)

        def predict_with(model):
            pred_df, _ = recursive_forward_predict_y(model, FEATS_Y, hist_min2[["ds","y","orders","stock"]], exog_selected, TEST_START, TEST_END)
            gt = test[["ds","y"]].merge(pred_df, on="ds", how="left")
            return gt

        test_rows=[]
        gt_rf = predict_with(rf_fit)
        m1, r1, p1 = mae_rmse_mape(gt_rf["y"], gt_rf["yhat"])
        test_rows.append({"Horizon":"Full(6m)", "Exog": best_exog_name, "Y-Variant":"RF", "MAE": m1, "RMSE": r1, "MAPE": p1})

        if HAS_XGB:
            gt_xgb = predict_with(xgb_fit)
            m2, r2, p2 = mae_rmse_mape(gt_xgb["y"], gt_xgb["yhat"])
            test_rows.append({"Horizon":"Full(6m)", "Exog": best_exog_name, "Y-Variant":"XGB", "MAE": m2, "RMSE": r2, "MAPE": p2})

        if HAS_XGB:
            yhat_ens = yens_w["RF"]*gt_rf["yhat"].values + yens_w["XGB"]*gt_xgb["yhat"].values
            m3, r3, p3 = mae_rmse_mape(gt_rf["y"], yhat_ens)
            test_rows.append({"Horizon":"Full(6m)", "Exog": best_exog_name, "Y-Variant":"Y-ENS", "MAE": m3, "RMSE": r3, "MAPE": p3, "w_RF": yens_w["RF"], "w_XGB": yens_w["XGB"]})

        test_summary = pd.DataFrame(test_rows)
        print("\n=== TEST Summary — SELECTED EXOG ===")
        print(test_summary.to_string(index=False))

        best_row = test_summary.sort_values("MAE").iloc[0]
        best_yvar = best_row["Y-Variant"]
        if best_yvar=="RF":
            gt_use = gt_rf
        elif best_yvar=="XGB" and HAS_XGB:
            gt_use = gt_xgb
        else:
            gt_use = gt_rf.copy()
            gt_use["yhat"] = yens_w["RF"]*gt_rf["yhat"].values + yens_w["XGB"]*(gt_xgb["yhat"].values if HAS_XGB else 0.0)

        resids = (gt_rf["y"].values - gt_rf["y"].values)  # placeholder init
        if best_yvar=="RF":
            resids = gt_rf["y"].values - gt_rf["yhat"].values
        elif best_yvar=="XGB" and HAS_XGB:
            resids = gt_xgb["y"].values - gt_xgb["yhat"].values
        else:
            resids = gt_rf["y"].values - (yens_w["RF"]*gt_rf["yhat"].values + yens_w.get("XGB",0.0)*(gt_xgb["yhat"].values if HAS_XGB else 0.0))

        sims, lo80, hi80, lo95, hi95 = smoothed_bootstrap_intervals(gt_use["yhat"].to_numpy(), resids, B=800, rng=rng_global)
        fc = gt_use[["ds","yhat"]].copy()
        fc["pi80_lo"]=np.maximum(0.0, lo80); fc["pi80_hi"]=np.maximum(0.0, hi80)
        fc["pi95_lo"]=np.maximum(0.0, lo95); fc["pi95_hi"]=np.maximum(0.0, hi95)

        cfg_row = cfg[cfg["sku"].astype(str)==str(sku)]
        H_COVER = int(cfg_row["H_COVER"].iloc[0]) if not cfg_row.empty else 6
        q_target= float(cfg_row["q_target"].iloc[0]) if not cfg_row.empty else 0.50
        moq     = int(cfg_row["MOQ"].iloc[0]) if not cfg_row.empty else 0
        lot     = int(cfg_row["lot_size"].iloc[0]) if not cfg_row.empty else 1

        start_idx = list(d["ds"]).index(TEST_START) if TEST_START in list(d["ds"]) else len(d)-1
        start_stock = float(d["stock"].iloc[max(0, start_idx-1)])
        p3, p6, e_t = stockout_stats(start_stock, sims)

        total_cov = sims[:min(6, H_COVER), :].sum(axis=0)
        q_total = float(np.nanquantile(total_cov, q_target))
        raw_order = max(0.0, q_total - start_stock)
        order_qty = round_moq_lot(raw_order, moq=moq, lot=lot)

        print("\n=== OMS Recommendation (policy-driven) ===")
        print(f"Selected combo: Exog={best_exog_name} | Y={best_yvar}")
        print(f"Starting stock: {start_stock:.2f}")
        print(f"Stockout: P<=3m={p3:.3f}, P<=6m={p6:.3f}, E[T_stockout]={e_t:.3f} months")
        print(f"Policy: cover {H_COVER}m @ q={q_target}")
        print(f"REORDER {'NEEDED' if order_qty>0 else 'NOT NEEDED'} → Order Qty ≈ {order_qty:.2f} (raw={raw_order:.2f})")

        out_csv = OUT_DIR/"forecasts"/f"{sku}__{best_exog_name}__{best_yvar}.csv"
        fc.to_csv(out_csv, index=False, date_format="%Y-%m-%d")

        hist12 = d[d["ds"] >= d["ds"].max() - pd.DateOffset(months=11)].copy()
        # 6m
        plt.figure(figsize=(10,5))
        plt.plot(hist12["ds"], hist12["y"], label="Actual (last 12m)")
        plt.plot(fc["ds"],    fc["yhat"], label="Forecast (6m)")
        plt.fill_between(fc["ds"], fc["pi80_lo"], fc["pi80_hi"], alpha=0.2, label="PI80")
        plt.fill_between(fc["ds"], fc["pi95_lo"], fc["pi95_hi"], alpha=0.1, label="PI95")
        plt.title(f"{sku} — 6m Forecast (EXOG: {best_exog_name} | Y: {best_yvar})")
        plt.xlabel("Date"); plt.ylabel("Sales"); plt.legend(); plt.tight_layout()
        plt.savefig(OUT_DIR/"plots"/f"{sku}__{best_exog_name}__{best_yvar}_6m.png", dpi=130)
        plt.show()

        # 3m
        plt.figure(figsize=(10,5))
        fc3 = fc.iloc[:3]
        plt.plot(hist12["ds"], hist12["y"], label="Actual (last 12m)")
        plt.plot(fc3["ds"],   fc3["yhat"], label="Forecast (3m)")
        plt.fill_between(fc3["ds"], fc3["pi80_lo"], fc3["pi80_hi"], alpha=0.2, label="PI80")
        plt.fill_between(fc3["ds"], fc3["pi95_lo"], fc3["pi95_hi"], alpha=0.1, label="PI95")
        plt.title(f"{sku} — 3m Forecast (EXOG: {best_exog_name} | Y: {best_yvar})")
        plt.xlabel("Date"); plt.ylabel("Sales"); plt.legend(); plt.tight_layout()
        plt.savefig(OUT_DIR/"plots"/f"{sku}__{best_exog_name}__{best_yvar}_3m.png", dpi=130)
        plt.show()

    print("\nTamamlandı. Çıktılar:", str(OUT_DIR))

if __name__ == "__main__":
    main()
```
